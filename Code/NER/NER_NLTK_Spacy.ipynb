```python
import nltk
# nltk.download('maxent_ne_chunker')
# nltk.download('words')
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
```

```python
sentence = 'Katong’s famous Hokkien mee and fried mee sua hawker stall Yong Huat at Alibabar kopitiam is closing permanently. Read more via link in our profile (click on the URL that starts with ‘linkinprofile’ below the 8 Days logo). 📷: Kelvin Chia bit.ly/2BPA4DS'
```

```python
ne_tree = ne_chunk(pos_tag(word_tokenize(sentence)))
```

```python
print(ne_tree)
```

```python
ex = sentence
```

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
sent = preprocess(ex)
sent
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
```

```python
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)
print(cs)
```

```python
NPChunker = nltk.RegexpParser(pattern) 
result = NPChunker.parse(sent)
result.draw()
```

```python
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint

iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
```

```python
ny_bb = url_to_string('https://www.nytimes.com/2018/08/13/us/politics/peter-strzok-fired-fbi.html?hp&action=click&pgtype=Homepage&clickSource=story-heading&module=first-column-region&region=top-news&WT.nav=top-news')
article = nlp(ny_bb)
len(article.ents)
```

```python
labels = [x.label_ for x in article.ents]
Counter(labels)
```

```python
items = [x.text for x in article.ents]
Counter(items).most_common(3)
```
