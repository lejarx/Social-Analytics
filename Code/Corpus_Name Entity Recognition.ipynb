# Name Entity Extraction with NLTK

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import conlltags2tree, tree2conlltags

from pprint import pprint
```

```python
poster_post[0]
```

```python
def preprocess(txt):
    txt = nltk.word_tokenize(txt)
    txt = nltk.pos_tag(txt)
    return txt
```

We shall now extract a list of tuples containing the words in the caption and ther associated part-of-speech.

```python
txt = poster_post[0]
txt_preprocessed = preprocess(txt)
txt_preprocessed
```

Chunk pattern rule where a noun phrase, NP, is formed when there is a optional determiner, DT, followed by any number of adjectives, JJ, and then a noun, NN.

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
```

```python
cp = nltk.RegexpParser(pattern)
cs = cp.parse(txt_preprocessed)
print(cs)
```

```python
iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
for x in iob_tagged:
    if x[2] != 'O':
        print(x)
```

# The following is testing out on Spacy

```python
#nlp = spacy.load("en_core_web_sm")
nlp = en_core_web_sm.load()
doc = nlp(poster_post)
new_corpus = []
for chunk in doc.noun_chunks:
    #print(chunk.text, chunk.root.text) #, chunk.root.dep_, chunk.root.head.text
    #df["food"] = np.char.add(chunk.root.text)
    new_corpus.append(chunk.root.text) #(chunk.text)

#df.head()
```

```python
posters_captions = []
for sentence in df["caption"]:
    posters_captions.append(sentence)

new_new_corpus = []
for each_sentence in posters_captions:
    for word in new_corpus:
        if each_sentence.find(word):
            new_new_corpus.append(word)
```
